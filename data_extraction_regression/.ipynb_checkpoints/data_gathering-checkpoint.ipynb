# Analysing time taken to extract block chain data

Block explorers either use a modified ethereum node or JSON-RPC calls to extract data from an ethereum node. Of these two methods, using JSON-RPC is an easier but time-consuming way to extract data. We hypothesize that if the optimizations in nodes remain same, as the usage and size of blockchain increases, catching up with the block chain for block explorers (which use JSON RPC) will become an increasingly difficult task. 

We will study two main effects:
1. Time taken to extract trace of transactions w.r.t. blockchain usage
2. Time taken to extract balance of 100 addresses w.r.t. blockchain size

```python
# importing basic libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from web3 import Web3, IPCProvider
from eth_utils import to_checksum_address
import time

w3 = Web3(IPCProvider(timeout=600))
# loading addresses with genesis rewards since they will be active throughout the blockchain history
df_genesis = pd.read_csv('genesis_rewards.csv', names=['address', 'balance'], nrows=100)
```

### Getting block meta data

```python
block_numbers = range(0, 5500000, 10000)
block_data_list = []
for i_block_number in block_numbers:
    block_data_list.append(w3.eth.getBlock(i_block_number))
# getting gas used per block
gas_used = [block_data['gasUsed'] for block_data in block_data_list]
```

###Â Calculating the time taken to fetch the address balance

```python
time_address_balance = []
for i_block_number in block_numbers:
    start_time = time.time()
    for index, row in df_genesis.iterrows():
        w3.eth.getBalance(to_checksum_address(row['address']), i_block_number)
    end_time = time.time() - start_time
    time_address_balance.append(end_time)
```

### Calculating the time taken to extract trace data

```python
def number_of_traces_in_block(trace):
    number_of_traces = 0
    for t in trace:
        number_of_traces += len(t)
    return number_of_traces

number_of_traces = []
time_trace = []
for i_block_number in block_numbers:
    start_time = time.time()
    trace = w3.parity.traceReplayBlockTransactions(i_block_number)
    end_time = time.time() - start_time
    time_trace.append(end_time)
    number_of_traces.append(number_of_traces_in_block(trace))
```

```python
# converting to dataframe for easier analysiss
d = {'block_numbers': block_numbers, \
      'gas_used': gas_used, \
      'time_address_balance': time_address_balance, \
      'time_trace': time_trace, \
      'number_of_traces': number_of_traces}
df = pd.DataFrame(data = d)
```

### Basic plotting

```python
f, axarr = plt.subplots(2, 2, figsize=(16, 10))


axarr[0, 0].plot(df['block_numbers'], df['time_trace'], 'ro')
axarr[0, 0].set_xlabel('Block number')
axarr[0, 0].set_ylabel('Time taken for trace extraction')

axarr[0, 1].plot(df['block_numbers'], df['time_address_balance'], 'bo')
axarr[0, 1].set_xlabel('Block number')
axarr[0, 1].set_ylabel('Time taken for address extraction')

axarr[1, 0].plot(df['block_numbers'], df['gas_used'], 'r*')
axarr[1, 0].set_xlabel('Block number')
axarr[1, 0].set_ylabel('Gas used')

axarr[1, 1].plot(df['block_numbers'], df['number_of_traces'], 'b*')
axarr[1, 1].set_xlabel('Block number')
axarr[1, 1].set_ylabel('Number of traces')

plt.show()
```

## Finding correlations

Some basic comments:
1. The average block time for the ethereum block-chain is ~15 seconds
2. The time taken to extract 100 addresses is not a limiting factor which is maximum in a range of ~2 seconds
3. The time taken to extract trace data (data with internal transactions of contracts) does exceed the 15 second mark, we would be interested in finding which variables effect this factor the most.
4. The ethereum blockchain between the blocks 2.3M to 2.47M suffered DOS attacks (https://edcon.io/2017/ppt/one/Martin%20Holst%20Swende_The%20'Shanghai%20'Attacks_EDCON.pdf), these block numbers are not optimal for finding correlations between gas_used and time taken. 

### Inspecting time taken to extract traces
We will now remove the data for DOS block numbers and find the correlations.

```python
df_filtered = df[(df.block_numbers > 2470000) | (df.block_numbers < 2300000)]
df_corr = df_filtered.corr()
df_corr['time_trace']
```

We can see that the time taken to extract traces is strongly correlated to the variables 'gas_used' and 'number_of_traces' in a block. Let us visualize these relations to get a better understanding.

```python
f, axarr = plt.subplots(2, 2, figsize=(16, 10))

axarr[0, 0].plot(df_filtered['gas_used'], df_filtered['time_trace'], 'ro')
axarr[0, 0].set_xlabel('Gas used')
axarr[0, 0].set_ylabel('Time taken for trace extraction')
axarr[0, 0].set_title('Filtered blocks')

axarr[0, 1].plot(df_filtered['number_of_traces'], df_filtered['time_trace'], 'bo')
axarr[0, 1].set_xlabel('Number of traces in block')
axarr[0, 1].set_ylabel('Time taken for trace extraction')
axarr[0, 0].set_title('Filtered blocks')

# Plotting for cases when gas_used > 7M for better clarity
df_gas_7M = df[(df.gas_used > 7000000)]
axarr[1, 0].plot(df_gas_7M['gas_used'], df_gas_7M['time_trace'], 'ro')
axarr[1, 0].set_xlabel('Gas used')
axarr[1, 0].set_ylabel('Time taken for trace extraction')
axarr[1, 0].set_title('Gas used > 7M')

axarr[1, 1].plot(df_gas_7M['number_of_traces'], df_gas_7M['time_trace'], 'bo')
axarr[1, 1].set_xlabel('Number of traces in block')
axarr[1, 1].set_ylabel('Time taken for trace extraction')
axarr[1, 1].set_title('Gas used > 7M')

plt.show()
```

#### Primary observations:
0. Both the variables are strongly correlated to the time to extract traces.
1. Even when gas used ~8 million, some block traces can be extracted in ~ 2 seconds while others can be extracted at ~ 16 seconds.
2. This huge disparity can be accounted by the number of traces produced within the block, or number of internal contract calls in the block.
3. There might be other variables which can explain further variations (like specific OPCODES), we will not go into that much detail for the current investigation.

Let, us now build a simple linear regression model to extrapolate the time taken variable.

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
regr = linear_model.LinearRegression()
X = df[['gas_used', 'number_of_traces']]
y = df[['time_trace']]
model = regr.fit(X, y)

# The intercept
print('Intercept:', model.intercept_)
# The coefficients
print('Coefficients:', model.coef_)
```

#### Primary inferences
1. An increase in gas used by 1 million units will result in an increase in trace time by 1.132 seconds
2. An increase in number of traces by 1000 units will result in an increase in trace time by 0.3 seconds
